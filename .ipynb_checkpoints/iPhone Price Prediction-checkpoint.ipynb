```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
df = pd.read_excel('Iphone Price Analysis.xlsx')
df.info()
```

```python
df
```

```python
x = df[['Initial Price', 'iPhone Age']]
x
```

```python
y = df[['Current Price']]
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 43)
regression_model = LinearRegression().fit(x_train, y_train)
```

```python
regression_model.score(x_test, y_test)
```

```python
    #Prediction For iPhone 12 Rate within an year.
    #iPhone 12 was launched in 13th October 2020.
iphone_age = 1
initial_price = 80000
current_price = regression_model.predict([[initial_price, iphone_age]])
drop_precent = ((initial_price-current_price)/initial_price)*100    
plt_df = pd.DataFrame()
plt_df['Price'] = initial_price, current_price[0][0]
plt_df['Year'] = 2020, 2021
plt_df.plot(kind = 'bar', x='Year', y='Price', color='green')
plt.title("For iPhone 12, "+str(round(float(drop_precent),2))+"% Price Drop expected within 1 year.", pad=20,size=15, loc='left')
plt.grid()
plt.show()
print('Price of iPhone12 in 2020: ₹'+str(initial_price),'\nPrice of iPhone 12 in 2021: ₹'+str(math.ceil(current_price[0][0])))
```
